```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from IPython.core.pylabtools import figsize
import pylab as pl
import seaborn as sns
```

```python
MonthlyAIRHSR=pd.read_csv("DATA.csv")
high1x=MonthlyAIRHSR["month"]
high1y=MonthlyAIRHSR["AIR1"]
plt.plot(high1x,high1y,color="sandybrown")
high2x=MonthlyAIRHSR["month"]
high2y=MonthlyAIRHSR["AIR2"]
plt.plot(high2x,high2y,color="lightskyblue")
plt.legend(labels=["107/11-108/05","108/11-109/05"])
plt.xlabel("月份",fontweight="bold")
plt.ylabel("旅客人數",fontweight="bold")
plt.title("機場月旅客人數",fontsize=15,fontweight="bold")
```

```python
plt.xticks(rotation =60)
DailyAIR=pd.read_csv("AirlineAndCase.csv")
Date=DailyAIR["日期"]
DailyEntry=DailyAIR["Entry"]
DailyExit=DailyAIR["Exit"]
CaseAddNum=DailyAIR["cases"]

Airline = plt.gca() 
Airline.plot(Date,DailyEntry,color="dodgerblue",label="總入境人數")
Airline.plot(Date,DailyExit,color="orange",label="總出境人數")

Airline.set_title("出入境人數與境外確診數之比較")
Airline.set_ylabel("出入境人數")
Airline.set_xlabel("日期")

ticker_spacing =5
Airline.xaxis.set_major_locator(ticker.MultipleLocator(ticker_spacing))
plt.figure(figsize(40,10))

plt.show()
```

```python
plt.xticks(rotation =45)
DailyTRAMRT=pd.read_csv("DailyTrain_MRT.csv")
Date=DailyTRAMRT["Date"]
DailyTRA=DailyTRAMRT["TRA"]

TRA = plt.gca() 
TRA.plot(Date,DailyTRA,color="darkorange",label="旅客人數")

TRA.set_title("台灣鐵路每日旅客人數", fontsize=40)
TRA.set_ylabel("旅客人數", fontsize=30)
TRA.set_xlabel("日期", fontsize=30) 
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(labels=["旅客人數"], fontsize=30)

ticker_spacing =5
TRA.xaxis.set_major_locator(ticker.MultipleLocator(ticker_spacing))
plt.figure(figsize(40,10))
plt.show()
```

```python
plt.xticks(rotation =45)
DailyTRAMRT=pd.read_csv("DailyTrain_MRT.csv")
Date=DailyTRAMRT["Date"]
DailyMRT=DailyTRAMRT["MRT"]

MRT = plt.gca() 
MRT.plot(Date,DailyMRT,color="darkorange",label="旅客人數")

MRT.set_title("台北捷運每日旅客人數", fontsize=40)
MRT.set_ylabel("旅客人數", fontsize=30)
MRT.set_xlabel("日期", fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(labels=["旅客人數"], fontsize=30)

ticker_spacing =5
MRT.xaxis.set_major_locator(ticker.MultipleLocator(ticker_spacing))
plt.figure(figsize(40,10))

plt.show()
```
